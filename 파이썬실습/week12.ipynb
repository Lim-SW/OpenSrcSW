```python
# 실습#1:서울의 일교차
import csv
f = open('weather.csv', 'r', encoding='cp949')
data = csv.reader(f)

next(data)
# 날짜, 지점, 평균기온,  최저기온, 최고기온
#  0     1       2          3         4
# 일교차 = 최고 - 최저 = 4 - 3 float

best = -999.0
best_d = ''
lowest = 999.0
lowest_d = ''
highest = -999.0
highest_d = ''

for row in data: # 모든 데이터 순회
    temp1 = float(row[4])-float(row[3]) # 일교차 계산
    if (temp1 > best): # 일교차가 제일 큰 값보다 크면
        best = round(temp1,2) # 가장 큰 일교차 저장
        best_d = row[0] # 가장 큰 일교차인 날짜 저장
        #print(best_d, best)
    
    temp2 = float(row[3]) # 최저기온이
    if(temp2<lowest): # 현재 저장된 최저기온보다 낮으면
        lowest = round(temp2,2) # 가장 낮은 최저기온 저장
        lowest_d = row[0] # 가장 낮은 최저기온안 날짜 저장
        #print(lowest_d, lowest)
    
    temp3 = float(row[4]) # 최고기온이
    if(temp2>highest): # 현재 저장된 최고기온보다 높으면
        highest = round(temp3,2) # 가장 높은 최고기온 저장
        highest_d = row[0] # 가장 높은 최고기온인 날짜 저장
        #print(highest_d, highest)

print('일교차가 가장 큰 날')
print("{}, ({}'C)".format(best_d, best))
print()
print('최저기온이 가장 낮은 날')
print("{}, ({}'C)".format(lowest_d, lowest))
print()
print('최고기온이 가장 높은 날')
print("{}, ({}'C)".format(highest_d, highest))
```

```python
# 실습#2:지하철 이용객
import csv
f = open('subway.csv', 'r', encoding='UTF8')
data = csv.reader(f)
next(data)
# 사용월, 호선명, 역ID,  지하철역, 승차승객수, 하차승객수
#   0       1      2        3          4           5
# 이용객수 = 승차승객수 + 하차승객수 = 4 + 3 int

hosun = {} # 딕셔너리 형태로 저장 (많은순)
hosun_r = {} # 딕셔너리 형태로 저장 (적은순)

for row in data:
    if not (row[1] in hosun): # 호선별로 각 딕셔너리에 중복되지 않도록 항목 생성 (많은순)
        zero = ['',0,'',0,'',0]
        hosun[row[1]] = zero
        
    if not (row[1] in hosun_r): # 호선별로 각 딕셔너리에 중복되지 않도록 항목 생성 (적은순)
        zero = ['',99999999,'',99999999,'',99999999]
        hosun_r[row[1]] = zero
        
    temp = int(row[4].replace(',','')) + int(row[5].replace(',','')) # 이용객수 계산
    
    if(hosun[row[1]][1]<temp): # 1위보다 크면 제일 큰 값
        hosun[row[1]][5] = hosun[row[1]][3] # 3위에 2위의 이용객수 저장
        hosun[row[1]][4] = hosun[row[1]][2] # 3위에 2위의 역이름 저장
        hosun[row[1]][3] = hosun[row[1]][1] # 2위에 1위의 이용객수 저장
        hosun[row[1]][2] = hosun[row[1]][0] # 2위에 1위의 역이름 저장
        hosun[row[1]][1] = temp # 1위에 이용객수 저장
        hosun[row[1]][0] = row[3] # 1위에 역이름 저장
    elif(hosun[row[1]][3]<temp): # 1위보다 작고 2위보다 크면
        hosun[row[1]][5] = hosun[row[1]][3] # 3위에 2위의 이용객수 저장
        hosun[row[1]][4] = hosun[row[1]][2] # 3위에 2위의 역이름 저장
        hosun[row[1]][3] = temp # 2위에 이용객수 저장
        hosun[row[1]][2] = row[3] # 2위에 역이름 저장
    elif(hosun[row[1]][5]<temp): # 2위보다 작고 3위보다 크면
        hosun[row[1]][5] = temp # 3위에 이용객수 저장
        hosun[row[1]][4] = row[3] # 3위에 역이름 저장
        
    if(hosun_r[row[1]][1]>temp): #위의 조건문이랑 다 같은데 비교 부등호를 반대로
        hosun_r[row[1]][5] = hosun_r[row[1]][3]
        hosun_r[row[1]][4] = hosun_r[row[1]][2]
        hosun_r[row[1]][3] = hosun_r[row[1]][1]
        hosun_r[row[1]][2] = hosun_r[row[1]][0]
        hosun_r[row[1]][1] = temp
        hosun_r[row[1]][0] = row[3]
    elif(hosun_r[row[1]][3]>temp):
        hosun_r[row[1]][5] = hosun_r[row[1]][3]
        hosun_r[row[1]][4] = hosun_r[row[1]][2]
        hosun_r[row[1]][3] = temp
        hosun_r[row[1]][2] = row[3]
    elif(hosun_r[row[1]][5]>temp):
        hosun_r[row[1]][5] = temp
        hosun_r[row[1]][4] = row[3]
        
hosun = sorted(hosun.items(), reverse=False) # 호선별 이름순으로 정렬하여 리스트에 담음
hosun_r = sorted(hosun_r.items(), reverse=False)

for i in range(len(hosun)): # 총 호선 수 만큼 (25개의 호선)
    print('============{}============'.format(hosun[i][0])) # 호선 이름 출력
    print('이용객이 가장 많은 역 순위')
    for j in range(0,6,2): # 3위까지 출력 
        print('{}.{} ({}명)'.format(int(j/2+1),hosun[i][1][j], hosun[i][1][j+1])) # 많은순 리스트에 접근
    print()
    print('이용객이 가장 적은 역 순위')
    for k in range(0,6,2): # 3위까지 출력
        print('{}.{} ({}명)'.format(int(j/2+1),hosun_r[i][1][k], hosun_r[i][1][k+1])) # 적은순 리스트에 접근
    print()
print('==============================')
```
