```python
# 실습 #2: 최대값 구하기
list = input("Numbers? ")
list = list.split(' ')
max=0

for i in list:
    temp = int(i)
    if max <= temp:
        max=temp

print('Max =',max)
```

```python
# 실습 #3: 암호화/복호화
def Amho():
    print('====================================')
    text = input('암호화할 문장: ')
    print('====================================')
    text = text.split(' ')
    text.reverse()
    tempT = ''
    for i in text:
        for j in i:
            temp = ord(j)+1
            tempT += chr(temp)
        tempT += ' '
    print('암호화된 문장:',tempT)

def Bokho():
    print('====================================')
    text = input('복호화할 문장: ')
    print('====================================')
    text = text.split(' ')
    text.reverse()
    tempT = ''
    for i in text:
        for j in i:
            temp = ord(j)-1
            tempT += chr(temp)
        tempT += ' '
    print('복호화된 문장:',tempT)
    
while(True):
    print('====================================')
    print('1.암호화 2.복호화 3.종료')
    inp = int(input())
    if inp == 1:
        Amho()
    elif inp == 2:
        Bokho()
    elif inp == 3:
        break
    else:
        continue
```

```python
def name():
    print('=========================================')
    text = input('찾으려는 전화번호의 이름: ')
    if di.get(text):
        print('=========================================')
        print(text,':',di[text])
    else:
        print('=========================================')
        print('조회되지 않습니다.')

def numb():
    print('=========================================')
    text = input('찾으려는 이름의 전화번호: ')
    if diR.get(text):
        print('=========================================')
        print(text,':',diR[text])
    else:
        print('=========================================')
        print('조회되지 않습니다.')

di = {}
for i in range(1,11):
    a = input(str(i)+'번째 이름을 입력하세요:')
    b = input(str(i)+'번째 전화번호를 입력하세요:')
    di[a]=b

diR = dict(zip(di.values(),di.keys()))

while(True):
    print('=========================================')
    print('1.이름으로찾기 2.전화번호로찾기 3.종료')
    inp = int(input())
    if inp == 1:
        name()
    elif inp == 2:
        numb()
    elif inp == 3:
        break
    else:
        continue
```
