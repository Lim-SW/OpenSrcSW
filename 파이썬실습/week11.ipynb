```python
# 실습 #6: copy
inputStr = input('cp 파일1 파일2 입력:')
inputStr = inputStr.split(' ')

if(len(inputStr)==3 and inputStr[0]=='cp'):
    file1 = open(inputStr[1],'r', encoding='UTF8')
    file2 = open(inputStr[2],'w', encoding='UTF8')
    for line in file1:
        file2.write(line)
    file1.close()
    file2.close()
    print('{0}를, {1}로 복사완료'.format(inputStr[1],inputStr[2]))

else:
    print('잘못입력하셨습니다.')
```

```python
# 실습 #7: wc
inputStr = input('wc 파일1 입력:')
inputStr = inputStr.split(' ')

if(len(inputStr)==2 and inputStr[0]=='wc'):
    file = open(inputStr[1],'r', encoding='UTF8')
    lineCount = int(0)
    wordCount = int(0)
    for line in file:
        lineCount += 1
        line = line.split(' ')
        if(line[0] != '\n'): #엔터친 공백 단어에 포함해서 예외처리
            wordCount += len(line)
    file.close()
    
    print('{0}의 라인 수:{1}'.format(inputStr[1], lineCount))
    print('{0}의 단어 수:{1}'.format(inputStr[1], wordCount))

else:
    print('잘못입력하셨습니다.')
```

```python
# 실습 #8: 사칙 연산 클래스
class Calc:
    def sum(self, a, b):
        result = a+b
        print('{0} + {1} = {2}입니다.'.format(a,b,result))
    
    def subtraction(self, a, b):
        result = a-b
        print('{0} - {1} = {2}입니다.'.format(a,b,result))
    
    def multiplication(self, a, b):
        result = a*b
        print('{0} * {1} = {2}입니다.'.format(a,b,result))
    
    def division(self, a, b):
        if(b==0):
            print('0으로 나눌수 없습니다.')
        else:
            result = a/b
            print('{0} / {1} = {2}입니다.'.format(a,b,result))

if __name__ == "__main__":
    calc = Calc()
    calc.sum(1,2)
    calc.subtraction(5,1)
    calc.multiplication(2,3)
    calc.division(6,2)
```

```python
# 실습 #9: 계산기
class Stack:
    def __init__(self): # 빈스택으로 초기화
        self.data = [] 
        
    def size(self): # 스택 크기 확인
        return len(self.data)
    
    def push(self, item): # push 구현
        self.data.append(item)
        
    def pop(self): # pop 구현
        return self.data.pop()
    
    def peek(self): # 스택의 top 확인
        return self.data[-1]

if __name__ == "__main__":
    s1 = Stack() # 숫자 스택
    s2 = Stack() # 연산자 스택

    prec = { # 연산자에 우선순위 부여
                '*' : 3,
                '/' : 3,
                '+' : 2,
                '-' : 2,
            }

    #cal = '1 + 2 - 3 * 4 + 5'
    inp = input('계산식 입력:') # 계산식 입력받음
    cal = inp.split(' ') # ' ' 단위로 슬라이스

    for i in range(len(cal)):
        if (cal[i]=='*' or cal[i]=='/' or cal[i]=='+' or cal[i]=='-'): # 연산자면
            s2.push(cal[i]) # 연산자 스택에 push
            #print(cal[i],'들어감')
        elif(i == len(cal)-1): # 마지막 반복이면
            s1.push(int(cal[i])) # 숫자이므로 무조건 숫자 스택에 일단 push
            #print(cal[i],'들어감')
            while(s2.size()>=1): # 남은 스택의 계산들 모두 처리 (연산자 없을때 까지)
                temp = s2.pop()
                if(temp=='+'):
                    a2 = s1.pop()
                    a1 = s1.pop()
                    result = a1+a2
                    #print(temp,result)
                    s1.push(result)
                elif(temp=='-'):
                    a2 = s1.pop()
                    a1 = s1.pop()
                    result = a1-a2
                    #print(temp,result)
                    s1.push(result)
                elif(temp=='*'):
                    a2 = s1.pop()
                    a1 = s1.pop()
                    result = a1*a2
                    #print(temp,result)
                    s1.push(result)
                else:
                    a2 = s1.pop()
                    a1 = s1.pop()
                    result = a1/a2
                    #print(temp,result)
                    s1.push(result)
            print('계산식 출력:{} = {}'.format(inp,s1.pop())) # 모두 계산된 마지막 숫자큐의 결과 pop해서 출력
        else: # 숫자면
            s1.push(int(cal[i])) # 일단 숫자 스택에 push
            #print(cal[i],'들어감')
            if(s1.size()>=2): # 숫자 스택에 2개 이상의 값이 쌓여있으면
                temp = s2.peek() # 들어가있는 연산자를 확인해서
                if(prec[temp]>=prec[cal[i+1]]): # 뒤에 올 연산자와 우선순위 비교 후 우선순위가 크면
                    temp = s2.pop()             # 계산함
                    if(temp=='+'):
                        a2 = s1.pop()
                        a1 = s1.pop()
                        result = a1+a2
                        #print(temp,result)
                        s1.push(result)
                    elif(temp=='-'):
                        a2 = s1.pop()
                        a1 = s1.pop()
                        result = a1-a2
                        #print(temp,result)
                        s1.push(result)
                    elif(temp=='*'):
                        a2 = s1.pop()
                        a1 = s1.pop()
                        result = a1*a2
                        #print(temp,result)
                        s1.push(result)
                    else:
                        a2 = s1.pop()
                        a1 = s1.pop()
                        result = a1/a2
                        #print(temp,result)
                        s1.push(result)         #계산하고 추가로 다음 연산자 또 확인해야함

                    if(cal[i+1] == '+' or cal[i+1] == '-'): # 다음 연산자가 +, -면 앞에 쌓인 값들을
                        while(s2.size()!=0):                # 모두 연산 처리 해줘야함
                            temp = s2.pop()                 # 그래야 계산 순서가 맞음
                            if(temp=='+'):
                                a2 = s1.pop()
                                a1 = s1.pop()
                                result = a1+a2
                                #print(temp,result)
                                s1.push(result)
                            elif(temp=='-'):
                                a2 = s1.pop()
                                a1 = s1.pop()
                                result = a1-a2
                                #print(temp,result)
                                s1.push(result)
                            elif(temp=='*'):
                                a2 = s1.pop()
                                a1 = s1.pop()
                                result = a1*a2
                                #print(temp,result)
                                s1.push(result)
                            else:
                                a2 = s1.pop()
                                a1 = s1.pop()
                                result = a1/a2
                                #print(temp,result)
                                s1.push(result)
                            
# 테스트 결과 복잡한(*, /가 연속되고 중간에 +,- 가 중간에 있는 식 포함) 식 모두 계산값이 정확함
# 계산식 입력:1 * 2 * 3 + 4 - 5 * 6 * 7 + 8
# 계산식 출력:1 * 2 * 3 + 4 - 5 * 6 * 7 + 8 = -192
```
