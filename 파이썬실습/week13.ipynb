```python
# 실습#1:생일 기온 시각화
import csv
import matplotlib.pyplot as plt

# 2011.01.01 ~ 2020.12.31 서울 기온 데이터
f = open('week13.csv', 'r', encoding='cp949')
data = csv.reader(f)

next(data)
# 날짜, 지점, 평균기온,  최저기온, 최고기온
#  0     1       2          3         4

high = [] # 최고 기온 누적 리스트
low = [] # 최저 기온 누적 리스트

for row in data:
    if row[-1] != '' : # 빈 데이터 스킵
        if (row[0].split('-')[1] == '06') and (row[0].split('-')[2] == '05'):
        # 월이 6월, 일이 5일 이면
            high.append(float(row[4])) # 최고기온 high에 append
            low.append(float(row[3])) # 최저기온 low에 append
            
#plt.title('06/05')
plt.plot(high, color='pink')#, label='HIGH')
# 강의자료와 동일하게 핑크색으로 그래프 출력
plt.plot(low, color='skyblue')#, label='LOW')
# 강의자료와 동일하게 하늘색으로 그래프 출력
plt.show() # 출력
```

```python
# 실습#2:최고 기온 시각화
import csv
import matplotlib.pyplot as plt

# 2011.01.01 ~ 2020.12.31 서울 기온 데이터
f = open('week13.csv', 'r', encoding='cp949')
data = csv.reader(f)

next(data)
# 날짜, 지점, 평균기온,  최저기온, 최고기온
#  0     1       2          3         4

jan = [] # 1월의 최고기온 데이터
aug = [] # 8월의 최고기온 데이터

for row in data:
    if row[-1] != '' : # 빈 데이터 스킵
        if(row[0].split('-')[1] == '01'): # 월이 1월이면
            jan.append(float(row[4])) # jan 리스트에 append
        elif(row[0].split('-')[1] == '08'): # 월이 8월이면
            aug.append(float(row[4])) # aug 리스트에 append

plt.hist(aug, bins=100, color='red', label='Aug')
# 히스토그램으로 1월최고기온리스트 인자값 넘김, (bins=100), 범례레이블=Aug
plt.hist(jan, bins=100, color='blue', label='Jan')
# 히스토그램으로 8월최고기온리스트 인자값 넘김, (bins=100), 범례레이블=Jan
plt.legend() # 범례(label) 적용
plt.show()
```
